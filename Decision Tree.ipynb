# DECISION TREE CLASSIFIER

#### Importing the required libraries

```python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Importing the dateset

```python
iris=datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

#### Splitting the data into train and test 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

#### Converting the values into Standard Values

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state = 0)
classifier.fit(X_train, y_train)
```

### Importing the libraries required for creating and displaying the Decision Tree

```python
from io import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
```

```python
dot_data = StringIO()
export_graphviz(classifier, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
```

#### Predicting the classes for Test Set

```python
y_pred = classifier.predict(X_test)
```

### Evaluating the Model Using Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

### Displaying the Confusion Matrix

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```

### Accuracy Score for Classification

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
